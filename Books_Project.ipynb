#1. Extract data 

1.1. Extract data from CSV files

```python
import pandas as pd
from google.colab import files
#upload=files.upload() 
df_books=pd.read_csv('BX-Books.csv', sep='delimeter', encoding='latin-1')
df_users=pd.read_csv('BX-Users.csv', sep='delimeter', encoding='latin-1')
df_ratings=pd.read_csv('BX-Book-Ratings.csv', sep='delimeter', encoding='latin-1')
```

#2. Transform data

2.1. Transform data in a format we can use

```python
# Function for transforming the data

def transform_data(df_initial,string_cols):
  df_rezult=pd.DataFrame()
  cols=df_initial.columns
  str_cols=str(cols)
  split_cols=str_cols.split('"')[1::2]
  nr=0
  for col in split_cols:
    df_rezult[f'{col}']=df_initial[f'{string_cols}'].str.split(';',expand=True)[nr]
    df_rezult[f'{col}']=df_rezult[f'{col}'].str.replace('"',"")
    nr=nr+1
  return df_rezult

# Books dataframe transformed
df_book=pd.DataFrame()
df_book=transform_data(df_books,'"ISBN";"Book-Title";"Book-Author";"Year-Of-Publication";"Publisher";"Image-URL-S";"Image-URL-M";"Image-URL-L"')
df_book['ISBN']=df_book['ISBN'].str.replace('X','')
display(df_book)
```

```python
# Users dataframe transformed
df_user=pd.DataFrame()
df_user=transform_data(df_users,'"User-ID";"Location";"Age"')
display(df_user)
```

```python
# Ratings dataframe transformed
df_rating=pd.DataFrame()
df_rating=transform_data(df_ratings,'"User-ID";"ISBN";"Book-Rating"')
df_rating['ISBN']=df_rating['ISBN'].str.replace('X','')
display(df_rating)
```

2.2. Filter only the books that have a rating over 8 and that have been published between 2000 and 2020

```python
# Merge dataset books with ratings 
df_books_rating=pd.merge(df_book,df_rating,on='ISBN')
df_books_rating=df_books_rating.drop_duplicates()
# Change data format
df_books_rating['Book-Rating']=pd.to_numeric(df_books_rating['Book-Rating'])
df_books_rating['Year-Of-Publication']=pd.to_numeric(df_books_rating['Year-Of-Publication'], errors='coerce')
df_books_rating=df_books_rating.dropna(subset='Year-Of-Publication')
df_books_rating['Year-Of-Publication']=df_books_rating['Year-Of-Publication'].astype(int)

# Filter books
df_wanted_books=df_books_rating.loc[(df_books_rating['Book-Rating']>8)&(df_books_rating['Year-Of-Publication']<2020)&(df_books_rating['Year-Of-Publication']>2000)]
display(df_wanted_books)
```

1.2. Extract data from Google Books api

```python
import requests

def search_books(query):
    api_key = 'AIzaSyDS9E2RrvaY1lG7ytJsmO363pvD06IW2qE'  # Replace with your actual API key

    # Send the request to the API
    url = f'https://www.googleapis.com/books/v1/volumes?q={query}&key={api_key}'
    response = requests.get(url)

    # Process the JSON response into a DataFrame
    data = response.json()
    books = data.get('items', [])

    # Extract relevant information from each book
    book_data = []
    for book in books:
        volume_info = book.get('volumeInfo', {})
        title = volume_info.get('title', 'N/A')
        authors = ', '.join(volume_info.get('authors', ['Unknown']))
        published_date = volume_info.get('publishedDate', 'N/A')
        categories = ', '.join(volume_info.get('categories', ['Unknown']))
        description= volume_info.get('description', 'N/A')
        identifiers = volume_info.get('industryIdentifiers', [])
        
        isbn = 'N/A'
        for identifier in identifiers:
            if identifier.get('type') == 'ISBN_13':
                isbn = identifier.get('identifier')
                break
            elif identifier.get('type') == 'ISBN_10':
                isbn = identifier.get('identifier')
                break


        book_data.append([isbn, title, authors, published_date, categories, description])
        

    # Create a pandas DataFrame
    df = pd.DataFrame(book_data, columns=['ISBN', 'Title', 'Authors', 'Published Date', 'Categories','Description'])
    return df

ls_aux=[]
df_book_nr=df_wanted_books.head(800)
for i in df_book_nr['Book-Title']:
  df_aux=pd.DataFrame()
  query=i
  df_aux=search_books(query)
  ls_aux.append(df_aux)

dff_book=pd.concat(ls_aux)  
display(dff_book)
```

2.3. Merge dataset containing books with dataset extracted from google books api

```python
df_merge_api=pd.merge(df_book, dff_book, on='ISBN')
df_merge_api=df_merge_api.drop_duplicates()
display(df_merge_api)

# Merge with wanted books
df_final_merge=pd.merge(df_merge_api, df_wanted_books, on='ISBN')

```

```python
import wikipediaapi

def fetch_wikipedia_data(page_titles):
    wiki_wiki = wikipediaapi.Wikipedia('en')  # Create a Wikipedia object

    data = []
    for title in page_titles:
        page = wiki_wiki.page(title)  # Fetch the page

        if page.exists():
            author = page.summary.split(' ')[0]  # Extract the first word as the author
            description = page.summary

            data.append([title, author, description])
        else:
            data.append([title, 'N/A', 'N/A'])

    df = pd.DataFrame(data, columns=['Title', 'Author', 'Description'])
    return df

# Example usage
lst=[]
df_book_nr2=df_wanted_books.head(5)
for i in df_book_nr2['Book-Title']:
  df_aux1=pd.DataFrame()
  query1=i
  df_aux1=fetch_wikipedia_data(query1)
  lst.append(df_aux1)

dff_authors=pd.concat(lst)  
display(dff_authors)

```
